````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import os

os.environ['CUDA_VISIBLE_DEVICES'] = '0,1' # è®¾ç½® GPU ç¼–å·ï¼Œå¦‚æœå•æœºå•å¡æŒ‡å®šä¸€ä¸ªï¼Œå•æœºå¤šå¡æŒ‡å®šå¤šä¸ª GPU ç¼–å·

# è®¾å®šä½¿ç”¨çš„è®¾å¤‡ä¸ºCUDAï¼Œå¦‚æœCUDAä¸å¯ç”¨ï¼Œå°†å›é€€åˆ°CPU
device = "cuda"

# æŒ‡å®šé¢„è®­ç»ƒæ¨¡å‹çš„è·¯å¾„
mode_name_or_path = './glm-4-9b-chat'

# ä»æŒ‡å®šè·¯å¾„åŠ è½½é¢„è®­ç»ƒçš„åˆ†è¯å™¨
tokenizer = AutoTokenizer.from_pretrained(mode_name_or_path, trust_remote_code=True)


# ä»æŒ‡å®šè·¯å¾„åŠ è½½é¢„è®­ç»ƒçš„æ¨¡å‹
# torch_dtypeæŒ‡å®šæ•°æ®ç±»å‹ï¼Œlow_cpu_mem_usageä¼˜åŒ–CPUå†…å­˜ä½¿ç”¨ï¼Œtrust_remote_codeå…è®¸åŠ è½½è¿œç¨‹ä»£ç 
model = AutoModelForCausalLM.from_pretrained(
    mode_name_or_path,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    trust_remote_code=True,
    device_map="auto"
).eval()  # å°†æ¨¡å‹è®¾ç½®ä¸ºè¯„ä¼°æ¨¡å¼
````

````python
# å®šä¹‰ç”¨æˆ·è¾“å…¥çš„æŸ¥è¯¢æ–‡æœ¬
query = "è¯·ä»‹ç»ä¸€ä¸‹AIå¤§æ¨¡å‹"

# ä½¿ç”¨åˆ†è¯å™¨çš„apply_chat_templateæ–¹æ³•æ¥å‡†å¤‡è¾“å…¥æ•°æ®
# è¿™ä¸ªæ–¹æ³•ä¼šæ ¹æ®èŠå¤©æ¨¡æ¿å°†ç”¨æˆ·è¾“å…¥æ ¼å¼åŒ–ä¸ºæ¨¡å‹å¯æ¥å—çš„æ ¼å¼
# add_generation_promptæ·»åŠ ç”Ÿæˆæç¤ºï¼Œtokenizeè¿›è¡Œåˆ†è¯ï¼Œreturn_tensorsæŒ‡å®šè¿”å›PyTorchå¼ é‡
# return_dictæŒ‡å®šè¿”å›å­—å…¸æ ¼å¼ï¼Œæ–¹ä¾¿åç»­å¤„ç†
inputs = tokenizer.apply_chat_template(
    [{"role": "user", "content": query}],
    add_generation_prompt=True,
    tokenize=True,
    return_tensors="pt",
    return_dict=True
)

# å°†è¾“å…¥æ•°æ®ç§»åŠ¨åˆ°æŒ‡å®šçš„è®¾å¤‡ä¸Š
inputs = inputs.to(device)
````

````python
# å®šä¹‰ç”Ÿæˆæ–‡æœ¬æ—¶çš„å‚æ•°
gen_kwargs = {
    "max_length": 2500,  # è®¾ç½®ç”Ÿæˆæ–‡æœ¬çš„æœ€å¤§é•¿åº¦
    "do_sample": True,  # æ˜¯å¦ä»å¯èƒ½çš„ä¸‹ä¸€ä¸ªè¯ä¸­éšæœºé€‰æ‹©
    "top_k": 1  # ä»æ¦‚ç‡æœ€é«˜çš„kä¸ªè¯ä¸­é€‰æ‹©
}

# ä½¿ç”¨torch.no_grad()ä¸Šä¸‹æ–‡ç®¡ç†å™¨æ¥ç¦ç”¨æ¢¯åº¦è®¡ç®—ï¼Œè¿™åœ¨æ¨ç†æ—¶å¯ä»¥å‡å°‘å†…å­˜ä½¿ç”¨
with torch.no_grad():
    # ä½¿ç”¨æ¨¡å‹çš„generateæ–¹æ³•ç”Ÿæˆæ–‡æœ¬
    outputs = model.generate(**inputs, **gen_kwargs)
    
    # æˆªå–ç”Ÿæˆçš„æ–‡æœ¬ï¼Œå»é™¤å¼€å¤´çš„æç¤ºéƒ¨åˆ†
    outputs = outputs[:, inputs['input_ids'].shape[1]:]
    
    # ä½¿ç”¨åˆ†è¯å™¨çš„decodeæ–¹æ³•å°†ç”Ÿæˆçš„è¯IDè§£ç å›æ–‡æœ¬ï¼Œå¹¶æ‰“å°å‡ºæ¥
    print(tokenizer.decode(outputs[0], skip_special_tokens=True))
````
